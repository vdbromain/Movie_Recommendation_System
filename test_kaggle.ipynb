```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("CSV_to_DB").getOrCreate()

meta = spark.read.csv("./data/kaggle/movies_metadata.csv", header=True, inferSchema=True)
```

```python
belongs_to_collection = meta.select('belongs_to_collection')

belongs_to_collection.write.json('belongs_to_collection.json')
```

```python
df_btc = spark.read.json("belongs_to_collection.json/")
```

```python
df_btc.show(1, truncate=False)
```
